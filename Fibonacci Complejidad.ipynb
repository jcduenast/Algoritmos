# Fibonacci

Desarrolle un programa iterativo que calcule el n termino de la  la serie de Fibonacci en Python y que retorne el tiempo de ejecución. Usando matplotlib gráficar el tiempo de ejecución en función de n.

```python
import matplotlib.pyplot as plt
import time
```

```python
times = [0, 0]
n = 30000

start = time.time()
a = 0
b = 1
c = a+b
for i in range (3,n+1):
    a = b
    b = c
    c = a+b
#     print "Iteración", i, "número", c
    temp = time.time() - start
    times.append(temp)
    
print (time.time() - start)
# print "Numero", i, "es", c
```

```python
plt.plot(times)
plt.axis([1,n,0,times[-1]])
plt.ylabel('Tiempo de ejecucion [seg]')
plt.xlabel('Numero de iteracion')
plt.show()
```
